# Utilities

```python
# function to measure_tflite_latency:
def measure_tflite_latency(tflite_model_path, sample_input, runs=50): # takes 3 arguments: model path, input sample, and number of runs
    interpreter = tf.lite.Interpreter(model_path=tflite_model_path) # load the TFLite model using the interpreter
    interpreter.allocate_tensors() # allocate memory for input and output tensors
    input_index = interpreter.get_input_details()[0]['index'] # get the index of the input tensor
    output_index = interpreter.get_output_details()[0]['index'] # get the index of the output tensor
    times = [] # list to store the inference times
    for _ in range(runs): # repeat inference 'runs' number of times
        start = time.time() # record start time before inference
        interpreter.set_tensor(input_index, sample_input) # set the input tensor with the sample input
        interpreter.invoke() # run inference
        _ = interpreter.get_tensor(output_index)
        times.append(time.time() - start) # calculate and store the inference time
    return np.mean(times) # return the average inference time over all runs

# function to convert_to_tflite_optimized:
def convert_to_tflite_optimized(model, save_path):  # takes a Keras model and a path to save the TFLite model
    converter = tf.lite.TFLiteConverter.from_keras_model(model)  # create a TFLite converter from the Keras model
    converter.optimizations = [tf.lite.Optimize.DEFAULT]  # enable default optimizations
    converter.target_spec.supported_types = [tf.float16]  # specify that float16 precision should be used for size reduction and faster inference
    tflite_model = converter.convert()  # convert the model to TensorFlow Lite format
    with open(save_path, 'wb') as f:  # open the output file in write-binary mode
        f.write(tflite_model)  # write the converted TFLite model to the file
    return os.path.getsize(save_path) / (1024 * 1024)  # return the file size in megabytes (MB)
```

# Trainig Model: CNN + Pruning


```python
# set the system
split_ratio = 0.20  # set the data split ratio to 0.10, 0.15, or 0.20
base_dir = '/content'  # path where the dataset is stored
metadata_path = os.path.join(base_dir, 'HAM10000_metadata.csv')  # full path to the metadata CSV file
img_dir_1 = os.path.join(base_dir, 'HAM10000_images_part_1')  # full path to the first folder of image files
img_dir_2 = os.path.join(base_dir, 'HAM10000_images_part_2')  # full path to the second folder of image files
```

```python
# load dataset
ham_df = pd.read_csv(metadata_path)  # load the metadata CSV file me
ham_df = shuffle(ham_df, random_state=42)  # shuffle the rows to ensure randomness
```

```python
# handling with imbalance class:
majority_class = 'nv'  # define the majority class label
df_major = ham_df[ham_df['dx'] == majority_class]  # create a DataFrame containing only the majority class samples
df_minor = ham_df[ham_df['dx'] != majority_class]  # create a DataFrame containing all the minority class samples

df_minor_upsampled = resample(  # upsample the minority classes to reduce imbalance
    df_minor,                   # the minority samples to be upsampled
    replace=True,              # sample with replacement
    n_samples=len(df_major)//2,  # number of samples to generate
    random_state=42            # fixed random seed for reproducibility
)

ham_df = pd.concat([df_major, df_minor_upsampled])  # combine the original majority class with the upsampled minority classes
ham_df = shuffle(ham_df, random_state=42).reset_index(drop=True)  # shuffle again and reset the index
```

```python
# function to find_image_path:
def find_image_path(image_id):  # define a function that takes an image ID and returns its full file path
    for folder in [img_dir_1, img_dir_2]:  # loop through the two image directories
        path = os.path.join(folder, image_id + '.jpg')  # construct the full path to the image file
        if os.path.exists(path): return path  # if the image file exists at that path, return it
    return None  # if the image is not found in either folder, return None


ham_df['image_path'] = ham_df['image_id'].apply(find_image_path)  # add a new column with the full image path for each image_id
ham_df = ham_df.dropna(subset=['image_path']).reset_index(drop=True)  # drop rows where the image was not found, and reset the index
```

```python
# split dataset
train_df, temp_df = train_test_split(  # split the dataset into training and temporary (val+test) sets
    ham_df,                            # the full filtered DataFrame
    test_size=split_ratio,            # the proportion of the data to reserve for validation and testing
    stratify=ham_df['dx'],            # ensure class distribution is preserved in both splits
    random_state=42                   # use a fixed seed for reproducibility
)

val_df, test_df = train_test_split(   # split the temporary set again into validation and test sets
    temp_df,                          # the temporary set from above
    test_size=0.5,                    # split equally between validation and test (50% each of temp_df)
    stratify=temp_df['dx'],          # maintain class distribution across both new splits
    random_state=42                  # fixed seed for reproducibility
)
```

# Data Augmentation

```python
img_height, img_width = 128, 128  # set the target image dimensions height and width in pixels
batch_size = 32  # define the number of images to be processed in each batch during training or evaluation

datagen_train = tf.keras.preprocessing.image.ImageDataGenerator(  # create a data generator for training images with augmentation
    rescale=1./255,               # normalize pixel values to the [0, 1] range
    rotation_range=30,            # randomly rotate images up to 30 degrees
    width_shift_range=0.1,        # randomly shift images horizontally by up to 10% of the width
    height_shift_range=0.1,       # randomly shift images vertically by up to 10% of the height
    shear_range=0.2,              # apply random shearing transformations
    zoom_range=0.2,               # randomly zoom in/out on images up to 20%
    brightness_range=[0.8, 1.2],  # randomly adjust brightness between 80% and 120%
    horizontal_flip=True          # randomly flip images horizontally
)
```

# Rescaling

```python
datagen_eval = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)  # create a data generator for validation/test without augmentation only rescaling

train_generator = datagen_train.flow_from_dataframe(  # create a training data generator using the training dataframe
    train_df,                      # dataFrame containing image paths and labels for training
    x_col="image_path",           # column name in the DataFrame that contains the image file paths
    y_col="dx",                   # column name in the DataFrame that contains the class labels
    batch_size=batch_size,        # number of images per batch
    class_mode="categorical",     # class labels will be one-hot encoded for multi-class classification
    target_size=(img_height, img_width),  # resize all images to this size
    shuffle=True                  # shuffle the training data for better training performance
)

val_generator = datagen_eval.flow_from_dataframe(  # create a validation data generator using the validation dataframe
    val_df,                        # dataFrame with validation image paths and labels
    x_col="image_path",           # column containing image paths
    y_col="dx",                   # column containing class labels
    batch_size=batch_size,        # number of images per batch
    class_mode="categorical",     # one-hot encode the labels
    target_size=(img_height, img_width),  # resize images to the model input size
    shuffle=False                 # do NOT shuffle — keeps order consistent for evaluation
)

test_generator = datagen_eval.flow_from_dataframe(  # create a test data generator using the test dataframe
    test_df,                       # dataFrame with test image paths and labels
    x_col="image_path",           # column containing image paths
    y_col="dx",                   # column containing class labels
    batch_size=batch_size,        # number of images per batch
    class_mode="categorical",     # one-hot encode the labels
    target_size=(img_height, img_width),  # resize images to the model input size
    shuffle=False                 # keeps results consistent for evaluation
)
```

# Optimized CNN Model




```python
# function to create a CNN model
def create_model():
    model = models.Sequential([  # Use a Sequential model (a linear stack of layers)

        # first convolutional block
        layers.SeparableConv2D(32, (3,3), activation='relu', input_shape=(img_height, img_width, 3)),
        # depthwise separable convolution with 32 filters and 3x3 kernel, ReLU activation
        # input shape corresponds to RGB images of size img_height x img_width
        BatchNormalization(),  # normalize the activations to stabilize and speed up training
        layers.MaxPooling2D((2,2)),  # downsample feature maps using 2x2 max pooling

        # second convolutional block
        layers.SeparableConv2D(64, (3,3), activation='relu'),  # separable conv layer with 64 filters
        BatchNormalization(),  # normalize activations
        layers.MaxPooling2D((2,2)),  # further downsample the feature maps

        # third convolutional block
        layers.SeparableConv2D(128, (3,3), activation='relu'),  # separable conv layer with 128 filters
        BatchNormalization(),  # normalize
        layers.MaxPooling2D((2,2)),  # max pooling to reduce spatial dimensions

        # fourth convolutional block
        layers.SeparableConv2D(128, (3,3), activation='relu'),  # another conv layer with 128 filters
        BatchNormalization(),  # normalize
        layers.MaxPooling2D((2,2)),  # final downsampling layer

        # global feature aggregation
        layers.GlobalAveragePooling2D(),  # convert 2D feature maps into a single vector by averaging

        # fully connected layer
        layers.Dense(256, activation='relu'),  # dense layer with 256 units and ReLU activation
        Dropout(0.3),  # dropout layer to reduce overfitting (30% dropout rate)

        # output layer
        layers.Dense(7, activation='softmax')  # final layer with 7 output units (one for each class), softmax for probabilities
    ])

    return model  # return the constructed model


model = create_model()  # create the CNN model using the previously defined create_model() function

optimizer = tf.keras.optimizers.Adam(learning_rate=5e-4)
# define the optimizer: Adam with a learning rate of 0.0005
# adam is a popular choice for training deep learning models due to its adaptive learning rates

loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
# define the loss function: Categorical Crossentropy with label smoothing
# label smoothing (0.1) helps prevent the model from becoming overconfident by softening the target labels

model.compile(
    optimizer=optimizer,  # use the Adam optimizer defined above
    loss=loss_fn,         # use the categorical crossentropy loss function with label smoothing
    metrics=['accuracy']  # track accuracy as the performance metric during training and evaluation
)
```

# Training Basiline Optimized CNN Model

```python
# baseline training
y_train = train_generator.classes  # get the class labels (as integers) for the training data
cw = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
# compute class weights to handle class imbalance
# 'balanced' mode gives higher weight to less frequent classes
class_weights = dict(enumerate(cw))  # convert the weights array into a dictionary format expected by Keras

early_stop = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
# callback to stop training early if validation loss doesn't improve for 8 epochs
# it also restores the best model weights before overfitting started

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
# callback to reduce learning rate by half if validation loss doesn't improve for 3 epochs
# minimum learning rate is set to 1e-6

tracker = EmissionsTracker(output_dir=base_dir, output_file=f"emissions_split{int(split_ratio*100)}.csv")
# create an emissions tracker to measure energy usage and CO2 emissions during training
tracker.start()  # start tracking emissions

start_time = time.time()  # record the start time for training duration measurement

history = model.fit(  # train the model
    train_generator,              # training data generator
    epochs=60,                    # maximum number of epochs
    validation_data=val_generator,  # validation data generator
    callbacks=[early_stop, reduce_lr],  # use early stopping and learning rate reduction
    class_weight=class_weights,  # apply computed class weights to handle class imbalance
    verbose=1                    # print progress bar during training
)

# save baseline model history
history_baseline_path = os.path.join(base_dir, f'history_baseline_split{int(split_ratio*100)}.npz')
# create a path to save the training history as a .npz file

np.savez(history_baseline_path,
         accuracy=history.history['accuracy'],            # save training accuracy history
         val_accuracy=history.history['val_accuracy'],    # save validation accuracy history
         loss=history.history['loss'],                    # save training loss history
         val_loss=history.history['val_loss'])            # save validation loss history
```

# Pruning Method with TFMOT

```python
prune_low_magnitude = tfmot.sparsity.keras.prune_low_magnitude  # load the pruning API from TensorFlow Model Optimization Toolkit (TFMOT)

# function for wraps layers to enable pruning of low-magnitude weights
pruning_params = {  # Define the pruning configuration parameters
    'pruning_schedule': tfmot.sparsity.keras.PolynomialDecay(  # use a polynomial decay schedule for sparsity
        initial_sparsity=0.0,   # start with 0% sparsity (no weights pruned initially)
        final_sparsity=0.5,     # gradually prune weights until reaching 50% sparsity
        begin_step=200,         # start pruning at training step 200
        end_step=1000           # finish pruning at training step 1000
    )
}

# modified clone_function to apply pruning to Layer instances
pruned_model = tf.keras.models.clone_model(  # create a new model by cloning the original model
    model,  # the original (pretrained) model
    clone_function=lambda layer: prune_low_magnitude(layer, **pruning_params)
    # for each layer, apply pruning if it matches certain conditions
    if isinstance(layer, (layers.SeparableConv2D, layers.Dense)) and layer.trainable
    # only apply pruning to trainable SeparableConv2D and Dense layers
    else layer  # Leave all other layers unchanged
)

pruned_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),  # use a lower learning rate for fine-tuning the pruned model
    loss=loss_fn,  # use the same loss function as before with label smoothing
    metrics=['accuracy']  # track accuracy during training
)

callbacks_pruning = [
    tfmot.sparsity.keras.UpdatePruningStep(),  # callback to update pruning step count during training
    early_stop,                                # early stopping to prevent overfitting
    reduce_lr                                  # reduce learning rate on plateau
]

tracker_pruned = EmissionsTracker(output_dir=base_dir, output_file=f"emissions_pruned_split{int(split_ratio*100)}.csv")
tracker_pruned.start()
start_time_pruned = time.time()

history_pruned = pruned_model.fit(
    train_generator, epochs=20,
    validation_data=val_generator,
    callbacks=callbacks_pruning,
    class_weight=class_weights,
    verbose=1
)
```

# Utilities

```python
# function to load history from a .npz file and convert it into a Pandas DataFrame
def load_history(path):
    if not os.path.exists(path):  # return None if file does not exist
        return None
    h = np.load(path)  # load the .npz file
    return pd.DataFrame({  # create a DataFrame from the saved metrics
        'accuracy': h['accuracy'],
        'val_accuracy': h['val_accuracy'],
        'loss': h['loss'],
        'val_loss': h['val_loss']
    })


# function to smooth curves using interpolation
def smooth_curve(x, y, kind='cubic', num_points=300):
    if len(x) < 2:  # If there are too few points, skip smoothing
        return x, y
    f = interp1d(x, y, kind=kind)  # create an interpolation function
    x_smooth = np.linspace(x.min(), x.max(), num_points)  # generate evenly spaced x values
    y_smooth = f(x_smooth)  # get smoothed y values using interpolation
    return x_smooth, y_smooth  # return smoothed x and y
```

# Evaluation Model

```python
# import necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.preprocessing import label_binarize
from scipy.interpolate import interp1d
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

#base directories and class setup
base_dir = '/content'
metadata_path = os.path.join(base_dir, 'HAM10000_metadata.csv')
img_dir_1 = os.path.join(base_dir, 'HAM10000_images_part_1')
img_dir_2 = os.path.join(base_dir, 'HAM10000_images_part_2')

class_names = ['akiec', 'bcc', 'bkl', 'df', 'mel', 'nv', 'vasc']  # class labels
n_classes = len(class_names)  # number of classes

# test data generator
def load_test_generator(split_ratio):
    # load metadata
    ham_df = pd.read_csv(metadata_path)

    # create image paths based on available folders
    ham_df['image_path'] = ham_df['image_id'].apply(
        lambda x: os.path.join(img_dir_1, x + '.jpg') if os.path.exists(os.path.join(img_dir_1, x + '.jpg'))
        else os.path.join(img_dir_2, x + '.jpg')
    )

    # drop entries where image path is missing
    ham_df = ham_df.dropna(subset=['image_path'])

    # simulate original training/test split
    _, temp_df = train_test_split(ham_df, test_size=split_ratio, stratify=ham_df['dx'], random_state=42)
    _, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['dx'], random_state=42)

    # create keras image generator for test set
    datagen = ImageDataGenerator(rescale=1./255)
    test_gen = datagen.flow_from_dataframe(
        test_df, x_col='image_path', y_col='dx', target_size=(128, 128),
        class_mode='categorical', batch_size=32, shuffle=False
    )

    return test_gen, test_df

# model evaluation function
def evaluate_model(model_path, test_gen, test_df, model_name="model", split=0.1):
    # load keras model from disk
    model = load_model(model_path, compile=False)

    # predict probabilities and get class indices
    preds = model.predict(test_gen, verbose=1)
    y_pred_idx = np.argmax(preds, axis=1)
    y_true_idx = test_gen.classes

    # compute main classification metrics
    accuracy = accuracy_score(y_true_idx, y_pred_idx)
    precision = precision_score(y_true_idx, y_pred_idx, average='weighted', zero_division=0)
    recall = recall_score(y_true_idx, y_pred_idx, average='weighted', zero_division=0)
    f1 = f1_score(y_true_idx, y_pred_idx, average='weighted', zero_division=0)

    # print metrics to console
    print(f"\n{model_name} metrics")
    print(f"accuracy:  {accuracy:.4f}")
    print(f"precision: {precision:.4f}")
    print(f"recall:    {recall:.4f}")
    print(f"f1-score:  {f1:.4f}")
    print("\nclassification report:")
    print(classification_report(y_true_idx, y_pred_idx, target_names=class_names))

    # plot confusion matrix
    cm = confusion_matrix(y_true_idx, y_pred_idx)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(f'confusion matrix - {model_name} split {int(split * 100)}%')
    plt.xlabel('predicted')
    plt.ylabel('true')
    conf_path = os.path.join(base_dir, f'confusion_matrix_{model_name.lower()}_split{int(split*100)}.png')
    plt.savefig(conf_path, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

    # plot classification report as bar chart
    report_dict = classification_report(y_true_idx, y_pred_idx, target_names=class_names, output_dict=True)
    report_df = pd.DataFrame(report_dict).transpose().iloc[:n_classes, :3]
    report_df.plot(kind='bar', figsize=(12, 8))
    plt.title(f'classification report - {model_name} split {int(split * 100)}%')
    plt.ylabel('score')
    plt.ylim(0, 1.05)
    plt.xticks(rotation=45)
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.legend(loc='lower right')
    plt.tight_layout()
    report_path = os.path.join(base_dir, f'report_plot_{model_name.lower()}_split{int(split*100)}.png')
    plt.savefig(report_path, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

    # return metrics for further comparison
    return accuracy, precision, recall, f1
```

```python
# define data splits and initialize results list
splits = [0.1, 0.15, 0.2]
results = []

# loop through each split to evaluate both models
for split in splits:
    print(f"\nevaluating split {split*100:.0f}%")

    # load test data for the current split
    test_gen, test_df = load_test_generator(split)

    # define model file paths
    baseline_model_path = os.path.join(base_dir, f'ham10000_baseline_model_split{int(split*100)}.h5')
    pruned_model_path = os.path.join(base_dir, f'ham10000_pruned_finetuned_model_split{int(split*100)}.h5')

    # evaluate baseline model
    acc_b, prec_b, rec_b, f1_b = evaluate_model(baseline_model_path, test_gen, test_df, "baseline", split)

    # evaluate pruned model
    acc_p, prec_p, rec_p, f1_p = evaluate_model(pruned_model_path, test_gen, test_df, "pruned", split)

    # store the results
    results.append([split, acc_b, prec_b, rec_b, f1_b, acc_p, prec_p, rec_p, f1_p])

# create a dataframe with the metrics
results_df = pd.DataFrame(results, columns=[
    'split', 'acc_baseline', 'prec_baseline', 'rec_baseline', 'f1_baseline',
    'acc_pruned', 'prec_pruned', 'rec_pruned', 'f1_pruned'
])

# define path and save metrics to csv
results_csv = os.path.join(base_dir, 'metrics_summary.csv')
results_df.to_csv(results_csv, index=False)

# print confirmation and display results
print(f"\nmetrics summary saved to: {results_csv}")
print(results_df)
```

# ROC Curve Visualization

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, roc_auc_score
from scipy.interpolate import interp1d

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# paths and class setup
base_dir = '/content'
metadata_path = os.path.join(base_dir, 'HAM10000_metadata.csv')
img_dir_1 = os.path.join(base_dir, 'HAM10000_images_part_1')
img_dir_2 = os.path.join(base_dir, 'HAM10000_images_part_2')

class_names = ['akiec', 'bcc', 'bkl', 'df', 'mel', 'nv', 'vasc']  # all 7 classes
n_classes = len(class_names)
splits = [0.1, 0.15, 0.2]  # data split ratios to evaluate

# test data generator function
def load_test_generator(split_ratio):
    df = pd.read_csv(metadata_path)

    # build full image paths based on which folder contains each image
    df['image_path'] = df['image_id'].apply(
        lambda x: os.path.join(img_dir_1, x + '.jpg') if os.path.exists(os.path.join(img_dir_1, x + '.jpg'))
        else os.path.join(img_dir_2, x + '.jpg')
    )

    # remove rows with missing image paths
    df = df.dropna(subset=['image_path'])

    # simulate the split used in training (no need to train again)
    _, temp_df = train_test_split(df, test_size=split_ratio, stratify=df['dx'], random_state=42)
    _, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['dx'], random_state=42)

    # define test image generator
    datagen = ImageDataGenerator(rescale=1./255)
    test_gen = datagen.flow_from_dataframe(
        test_df, x_col='image_path', y_col='dx', target_size=(128, 128),
        class_mode='categorical', batch_size=32, shuffle=False
    )

    return test_gen

# create and overlay ROC curve for a model
plt.figure(figsize=(10, 8))  # create one big figure for all curves

def add_roc_curve(model, test_gen, model_name, split):
    preds = model.predict(test_gen)  # model predictions
    y_true = test_gen.classes  # true labels
    y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))  # binarize for ROC calculation

    # compute ROC values flattened
    fpr_raw, tpr_raw, _ = roc_curve(y_true_bin.ravel(), preds.ravel())
    auc_score = roc_auc_score(y_true_bin, preds, average='macro')  # macro average AUC

    # remove duplicate FPR values for interpolation
    _, unique_idx = np.unique(fpr_raw, return_index=True)
    fpr_clean = fpr_raw[unique_idx]
    tpr_clean = tpr_raw[unique_idx]

    # interpolate to make smoother curves
    if len(fpr_clean) >= 3:
        interp_f = interp1d(fpr_clean, tpr_clean, kind='linear', fill_value='extrapolate')
        fpr = np.linspace(0, 1, 300)
        tpr = interp_f(fpr)
    else:
        fpr, tpr = fpr_raw, tpr_raw

    # plot the ROC curve for the model
    label = f'{model_name} ({int(split*100)}%) AUC={auc_score:.2f}'
    plt.plot(fpr, tpr, label=label, linewidth=2)

# loop through each split and evaluate both models
for split in splits:
    print(f"\n-processing {int(split*100)}% split")
    test_gen = load_test_generator(split)

    baseline_path = os.path.join(base_dir, f'ham10000_baseline_model_split{int(split*100)}.h5')
    pruned_path = os.path.join(base_dir, f'ham10000_pruned_finetuned_model_split{int(split*100)}.h5')

    model_baseline = load_model(baseline_path, compile=False)
    model_pruned = load_model(pruned_path, compile=False)

    add_roc_curve(model_baseline, test_gen, 'cnn', split)
    add_roc_curve(model_pruned, test_gen, 'pruned', split)

# finalize and save ROC plot
plt.plot([0, 1], [0, 1], 'k--', label='random')  # diagonal reference line
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('roc curves - cnn vs pruned')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='lower right')
plt.tight_layout()

roc_path = os.path.join(base_dir, 'roc_curves_all_splits.png')
plt.savefig(roc_path, dpi=300, bbox_inches='tight')
plt.show()

print(f"roc curves saved to: {roc_path}")
```
